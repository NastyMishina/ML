```python
data = pd.read_csv('https://raw.githubusercontent.com/koroteevmv/ML_course/2024/ML4.1%20metrics/heart.csv')
```

```python
y = data["output"]
X = data.drop("output", axis=1)
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC, SVR
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_transform = PolynomialFeatures(degree=2)
poly_transform.fit(X)
X_transform = poly_transform.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_transform, y, train_size=0.8)
```

```python
models = {"Логистическая регрессия": LogisticRegression(max_iter=10000),
          "Метод опорных векторов": SVC(),
          "Перцептрон": MLPClassifier(),
          "Деревья решений": DecisionTreeClassifier(),
          "Метод k ближайших соседей": KNeighborsClassifier(),
          "Гауссовский Наивный Байес": GaussianNB()
          }
table = pd.DataFrame()
```

```python
def learn_model(models, x_train, x_test, y_train, y_test):
  global table
  for name, model in models.items():
    model.fit(x_train, y_train)

    y_test_pred = model.predict(x_test)

    metrics = pd.DataFrame({
          'Accuracy': accuracy_score(y_test, y_test_pred),
          'f1 score': f1_score(y_test, y_test_pred),
          'precision score': precision_score(y_test, y_test_pred),
          'recall score': recall_score(y_test, y_test_pred)
          },
          index=[name]
      )
    table = pd.concat([table, metrics])
    print(name)
    print(confusion_matrix(y_test, y_test_pred), '\n')
  return table
```

```python
learn_model(models, x_train, x_test, y_train, y_test)
```

3. Повторите анализ для датасета, предназначенного для решения задачи регрессии. Используйте все метрики качества регрессии, изученные на лекции. Постройте 5 - 10 разных моделей регрессии.

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_squared_log_error, mean_absolute_percentage_error, median_absolute_error,max_error
```

```python
data_3 = fetch_openml('analcatdata_supreme')
print(data_3.DESCR)
```

```python
X_3, y_3 = data_3.data, data_3.target
```

```python
X_3.describe()
```

```python
x_train_3, x_test_3, y_train_3, y_test_3 = train_test_split(X_3, y_3, train_size=0.8)
```

```python
regression_models = {"Линейная регрессия": LinearRegression(),
          "Перцептрон": MLPRegressor(),
          "Деревья решений": DecisionTreeRegressor(),
          "Метод k ближайших соседей": KNeighborsRegressor(),
         "Случайный лес": RandomForestRegressor(),
          "Метод опорных векторов": SVR()
          }

table_regression = pd.DataFrame()
```

```python
def learn_reg_model(models, x_train, x_test, y_train, y_test):
  global table_regression
  for name, model in models.items():
    model.fit(x_train, y_train)

    y_test_pred = model.predict(x_test)

    metrics = pd.DataFrame({
          'r2': r2_score(y_test, y_test_pred),
          'MAE': mean_absolute_error(y_test, y_test_pred),
          'MSE': mean_squared_error(y_test, y_test_pred),
          'MSLE': mean_squared_log_error(y_test, y_test_pred),
          'MAPE': mean_absolute_percentage_error(y_test, y_test_pred),
          'MedianAS': median_absolute_error(y_test, y_test_pred),
           'MAX Error': max_error(y_test, y_test_pred)
          },
          index=[name]
      )
    table_regression = pd.concat([table_regression, metrics])
  return table_regression
```

```python
result_table = learn_reg_model(regression_models, x_train_3, x_test_3, y_train_3, y_test_3)

poly_transform_3 = PolynomialFeatures(degree=2)
poly_transform_3.fit(X_3)
X_3_transform = poly_transform_3.transform(X_3)

x_train_3t, x_test_3t, y_train_3t, y_test_3t = train_test_split(X_3_transform, y_3, train_size=0.8)

poly_linear_reg = {'Полиномиальная регрессия': LinearRegression()}

table_reg = learn_reg_model(poly_linear_reg, x_train_3t, x_test_3t, y_train_3t, y_test_3t)

pd.concat([table_reg, result_table])
```
