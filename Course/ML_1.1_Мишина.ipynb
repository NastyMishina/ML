4. Модифицируйте алгоритм градиентного спука таким образом, чтобы он останавливаля, если ошибка перестает снижаться. Для этого сравнивайте, на сколько понизилась ошибка на каждом шаге и если это снижение меньше некоторой заранее заданной очень маленькой величины - выходите из цикла. Эту величину можно взять одну миллионную, например.

```python
class Model2():
    """Модель парной линейной регрессии"""
    def __init__(self, alpha = 0, min_decrease = 0):
        self.b0 = 0
        self.b1 = 0
        self.alpha = alpha
        self.min_decrease = min_decrease
        
    def predict(self, x):
        return self.b0 + self.b1 * x # функция гипотезы
    
    def error(self, X, Y):
        return sum((self.predict(X) - Y)**2) / (2*len(X)) # Функция ошибки
    
    def fit(self, x, y, itteration):
        for _ in range(itteration):
            old_er = self.error(x, y)
            self.b0 -= self.alpha * sum(self.predict(x) - y) / len(x)
            self.b1 -= self.alpha * sum((self.predict(x) - y) * x) / len(x)    # изменение параметров функции гипотезы
            if self.error(x, y) > old_er:    # проверка на увеличение значения ошибки
                self.alpha /= 2   # уменьшение значения скорости обучения
            elif abs(old_er - self.error(x,y)) < self.min_decrease:   # проверка на существенное изменение ошибки
                break
```

```python
class Model3():
    """Модель парной линейной регрессии"""
    def __init__(self, alpha = 0, min_decrease = 0):
        self.b0 = 0
        self.b1 = 0
        self.alpha = alpha
        self.min_decrease = min_decrease
        
    def predict(self, x):
        return self.b0 + self.b1 * x # функция гипотезы
    
    def error(self, X, Y):
        return sum((self.predict(X) - Y)**2) / (2*len(X)) # Функция ошибки
    
    def fit(self, x, y, itteration):
        for i in range(itteration): # магическая переменная, лучше писать в поле класса
            old_er = self.error(x, y)
            self.b0 -= self.alpha * sum(self.predict(x) - y) / len(x)
            self.b1 -= self.alpha * sum((self.predict(x) - y) * x) / len(x)    # изменение параметров функции гипотезы
            if self.error(x, y) > old_er:    # проверка на увеличение значения ошибки
                self.alpha /= 2   # уменьшение значения скорости обучения
            elif abs(old_er - self.error(x,y)) < self.min_decrease:   # проверка на существенное изменение ошибки
                print(f'Изменения ошибки не существенны: {old_er - self.error(x,y)}. Алгоритм завершился на {i + 1} итерации.')
                break
        else:
            print('Достигнуто максимальное количество шагов')
        
```
